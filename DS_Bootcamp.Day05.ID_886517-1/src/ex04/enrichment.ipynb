```python
import pandas as pd
import numpy as np
import requests
```

```python
df=pd.read_json('../data/auto.json', orient='records')
df.head()
```

```python
pd.options.display.float_format='{:.2f}'.format
```

```python
sam=df.sample(n=200, random_state=21)
sam.head()
```

```python
d_fines=df['Fines']
sam['Fines']=np.random.choice(d_fines, size=len(sam))
d_refund=df['Refund']
sam['Refund']=np.random.choice(d_refund, size=len(sam))
```

```python
concat_rows=pd.concat([df, sam], ignore_index=True)
concat_rows
```

```python
np.random.seed(21)
data_year=np.random.randint(1980, 2019, size=len(concat_rows))
Year=pd.Series(data_year, name='Year')
Year
```

```python
fines=pd.concat([concat_rows,Year], axis=1)
fines
```

```python
surnames=pd.read_json('../data/surname.json', orient='values')
surnames.columns=surnames.iloc[0]
surnames.drop(0, inplace=True)
surnames.head()
```

```python
unique_nums=concat_rows['CarNumber'].unique()
np.random.seed(21)
sam_surnames=surnames['NAME'].sample(n=len(unique_nums), replace=True, ignore_index=True, random_state=21)
surname_series=pd.Series(sam_surnames)
surname_series
```

```python
owners=pd.DataFrame({'CarNumber':unique_nums, 'SURNAME':surname_series})
owners
```

```python
new_obs={
    'CarNumber': ['53YRNKRUS', 'ERSG563RUS', '2335589dRUS', 'GERFNOFNRUS', 'EFWEFWE4554RUS'],
    'Refund': [2, 1, 1, 2, 1],
    'Fines': [5899.00, 34566.00, 23245.00, 7890.00, 500.00],
    'Make': ['Volvo', 'Toyota', 'Ford', 'Mazda', 'Suzuki'],
    'Model': ['xc90', 'Corolla', 'Focus', 'cx5', 'vitara'],
    'Year': [2023, 2011, 1998, 2010, 2005]}

new_obs=pd.DataFrame(new_obs)
fines=pd.concat([fines, new_obs], ignore_index=True)
fines
```

```python
owners_updated=owners.iloc[:-20]
new_owner = {
    'CarNumber': ['45847GRGR9RUS', 'RFRGKL#$%RUS', 'GERGHGJ45RUS'],
    'SURNAME': ["JORDAN", "IVANOV", "RGONGO"]
}

new_owner=pd.DataFrame(new_owner)
owners_updated=pd.concat([owners_updated, new_owner], ignore_index=True)
owners_updated
```

```python
new_df1=pd.merge(left=fines, right=owners_updated, on='CarNumber', how='inner')
new_df1
```

```python
new_df2=pd.merge(left=fines, right=owners_updated, on='CarNumber', how='outer')
new_df2
```

```python
new_df3=pd.merge(left=fines, right=owners_updated, on='CarNumber', how='left')
new_df3
```

```python
new_df4=pd.merge(left=fines, right=owners_updated, on='CarNumber', how='right')
new_df4
```

```python
pt_fines=pd.pivot_table(fines, columns='Year', values='Fines', index=['Make', 'Model'], aggfunc={'Fines':['sum']})
pt_fines
```
