```python
import pandas as pd
import gc
```

```python
df=pd.read_csv('../data/fines.csv')
df.head()
```

```python
%%timeit
res=[]
for i in range(0,len(df)):
    res.append((df.iloc[i]['Fines']/df.iloc[i]['Refund'])*df.iloc[i]['Year'])

df['res_loop']=res
```

```python
%%timeit
res=[]
for index, row in df.iterrows():
    res.append((row['Fines']/row['Refund'])*row['Year'])

df['res_iterrows']=res
```

```python
%%timeit
df['res_apply']=df.apply(lambda x: (x['Fines']/x['Refund'])*x['Year'], axis=1)
```

```python
%%timeit
df['res_series']=(df['Fines']/df['Refund'])*df['Year']
```

```python
%%timeit
df['res_values']=(df['Fines'].values/df['Refund'].values)*df['Year'].values
```
